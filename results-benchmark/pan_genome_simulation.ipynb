```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(1,1000)

pan = 4500*(X**(0.3))

core = 4500*(X**(-0.2))

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(6,4))

ax.plot(X,pan,label="pan-genome")
ax.plot(X,core, label="core genome")

ax.set_ylabel("size")
ax.set_xlabel("number of genomes")

ax.legend(loc='best')
plt.show()

fig.savefig('pan-core.png')
```
